# **Mapear EDAD (arriba y abajo de la media)**

```python
#Mapear EDAD (arriba y abajo de la media)
#se calcula la media (promedio) de la columna EDAD del DataFrame estudiantes.
#Crea una nueva columna llamada 'EDAD_Mapeada'.
#Usa el método .apply() sobre la columna 'EDAD' que aplica una función a cada valor (x) de esa columna.
media_edad = estudiantes['EDAD'].mean()
estudiantes['EDAD_Mapeada'] = estudiantes['EDAD'].apply(lambda x: 'Arriba de la media' if x > media_edad else ('Abajo de la media' if x < media_edad else 'En la media'))
#Muestra el DataFrame con la nueva columna 'EDAD_Mapeada' ya agregada, para visualizar los resultados.
display(estudiantes)
```

```python


# Aplicar map para restar la media de cada valor
estudiantes["EDAD"].map(lambda x: x - media_edad)
```
