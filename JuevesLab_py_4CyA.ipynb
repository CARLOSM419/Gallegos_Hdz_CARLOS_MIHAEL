```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Cargar datos
df = pd.read_csv('Automobile_data.csv')

print(" MANIPULANDO LOS DATOS:")
print("="*50)

# 1. Limpiar datos problemáticos (los que tienen '?')
print("1. LIMPIANDO DATOS CON '?':")
df_clean = df.replace('?', np.nan)

# Convertir columnas numéricas
numeric_columns = ['price', 'horsepower', 'engine-size', 'curb-weight']
for col in numeric_columns:
    if col in df_clean.columns:
        df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce')

print(" Datos limpiados")

# 2. Crear nuevas columnas
print("\n2. CREANDO NUEVAS COLUMNAS:")
if 'horsepower' in df_clean.columns and 'engine-size' in df_clean.columns:
    df_clean['power-to-weight'] = df_clean['horsepower'] / df_clean['engine-size']
    print(" Columna 'power-to-weight' creada")

# 3. Filtrar datos
print("\n3. FILTRANDO DATOS:")
# Automóviles con precio mayor a $20,000
if 'price' in df_clean.columns:
    caros = df_clean[df_clean['price'] > 20000]
    print(f" Autos caros (>$20K): {len(caros)} autos")

# 4. Agrupar datos
print("\n4. AGRUPANDO POR MARCA:")
if 'make' in df_clean.columns and 'price' in df_clean.columns:
    precio_por_marca = df_clean.groupby('make')['price'].mean().sort_values(ascending=False)
    print(" Precio promedio por marca:")
    print(precio_por_marca.head())

# 5. Ordenar datos
print("\n5. ORDENANDO POR PRECIO:")
if 'price' in df_clean.columns:
    df_ordenado = df_clean.sort_values('price', ascending=False)
    print(" Top 5 autos más caros:")
    print(df_ordenado[['make', 'price']].head())

# 6. Estadísticas básicas
print("\n6. ESTADÍSTICAS:")
if 'price' in df_clean.columns:
    print(f" Precio - Min: ${df_clean['price'].min():.2f}")
    print(f" Precio - Max: ${df_clean['price'].max():.2f}")
    print(f" Precio - Promedio: ${df_clean['price'].mean():.2f}")

# VISUALIZACIONES
print("\n CREANDO GRÁFICAS...")
plt.figure(figsize=(15, 10))

# Gráfica 1: Precios por marca
plt.subplot(2, 2, 1)
if 'make' in df_clean.columns and 'price' in df_clean.columns:
    df_clean.groupby('make')['price'].mean().sort_values().tail(10).plot(kind='barh')
    plt.title('Top 10 Marcas Más Caras')
    plt.xlabel('Precio Promedio')

# Gráfica 2: Distribución de precios
plt.subplot(2, 2, 2)
if 'price' in df_clean.columns:
    df_clean['price'].hist(bins=20)
    plt.title('Distribución de Precios')
    plt.xlabel('Precio')
    plt.ylabel('Frecuencia')

# Gráfica 3: Tipo de combustible
plt.subplot(2, 2, 3)
if 'fuel-type' in df_clean.columns:
    df_clean['fuel-type'].value_counts().plot(kind='pie', autopct='%1.1f%%')
    plt.title('Distribución por Tipo de Combustible')

# Gráfica 4: Caballos de fuerza vs Precio
plt.subplot(2, 2, 4)
if 'horsepower' in df_clean.columns and 'price' in df_clean.columns:
    plt.scatter(df_clean['horsepower'], df_clean['price'])
    plt.title('Caballos de Fuerza vs Precio')
    plt.xlabel('Caballos de Fuerza')
    plt.ylabel('Precio')

plt.tight_layout()
plt.show()

print("\n ANÁLISIS COMPLETADO!")
print(f" Dataset original: {df.shape}")
print(f" Dataset limpio: {df_clean.shape}")
```
