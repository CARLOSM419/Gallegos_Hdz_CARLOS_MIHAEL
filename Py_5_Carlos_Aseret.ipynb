```python
# Importar pandas
import pandas as pd

# Cargar el archivo
reviews = pd.read_csv('/content/harvard_reviews.csv')

# ===== HISTORIA DE LOS DATOS =====
print(" HISTORIA DE LOS DATOS\n")

# usando iloc (selecci√≥n por posici√≥n)

print(reviews.iloc[0:3][["published_date", "rating", "title"]])
print()

# Seleccionar  (posici√≥n 4)
print("üîπ Una rese√±a espec√≠fica (fila 4 con iloc):")
print(reviews.iloc[4])
print()

# (por condici√≥n)
print("üîπ Rese√±as con calificaci√≥n perfecta (5 estrellas):")
perfectas = reviews.loc[reviews["rating"] == 5, ["published_date", "published_platform", "title"]]
print(perfectas.head())
print()

#  publicadas desde m√≥viles
print("üîπ Opiniones publicadas desde celular:")
moviles = reviews.loc[reviews["published_platform"] == "Mobile", ["title", "rating"]]
print(moviles.head())
print()

#  texto nulo o vac√≠o
print("üîπ Rese√±as sin texto (isnull):")
vacias = reviews.loc[reviews["text"].isnull()]
print(vacias if not vacias.empty else "No hay rese√±as vac√≠as")
print()

# ===== HISTORIA  =====

print(f"Total: {len(reviews)}")
print(f"Promedio de calificaci√≥n: {reviews['rating'].mean():.2f}")
print(f"Publicaciones desde m√≥vil: {(reviews['published_platform'] == 'Mobile').sum()}")
print(f"Publicaciones desde escritorio: {(reviews['published_platform'] == 'Desktop').sum()}")
```
